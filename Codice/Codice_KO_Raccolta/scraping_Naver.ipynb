```python
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
PATH='C:\\Users\\Fossati\\Desktop\\Tesi_Ale\\chromedriver91.exe'
driver = webdriver.Chrome(PATH)
```

```python
# liste ad hoc 
title=meta_ko['Title'].to_list()
isbn=meta_ko['ISBN'].to_list()
gen=meta_ko['Genere'].to_list()
booklist=meta_ko['Naver_Link'].to_list()

nrev_list=meta_ko['Num_Reviews_Naver'].to_list()
print(booklist[:10])
```

```python
#componiamo una nuova lista link indirizzata direttamente sulle reviews

start="https://book.naver.com/bookdb/review.naver?bid="
my_links=[]
for link in booklist:
    cod=re.findall(r'\d+' ,str(link))
    my_links.append(str(start) + str(cod[0]))
print(len(my_links))
my_links[:10]
```

```python
#queste altre liste servono per la seconda parte dedicata alle reviews specifiche:

#riutilizziamo le liste titles e isbn create sopra
rev_content=[]
ratings=[]
rev_date=[]
userlist=[]
#number_review=[]

n_likes=[]
#vanno fillate ogn i volta
titoli=[]
isbn_code=[]
genere=[]
```

```python
#CATTURA REVIEWS NAVER


not_found=[]
number=-1
for link in my_links[:200]:
    
    linkrev=[] #catturiamo i link delle reviews per ogni libro e le catturiamo alla fine del for principale
    number+=1
    
    try:
        try:
            url= str(link)
            driver.get(url)
        except:
            time.sleep(5)
            url= str(link)
            driver.get(url)
    except:
        
        not_found.append(number)
        time.sleep(2)
        continue
    
    #n rev
    
    try:
        
            time.sleep(1)
            rat_rev=driver.find_element_by_xpath("//div[@class='book_info_inner']/div[@class='txt_desc']/a[@id='txt_desc_point']").text
            rr_list=re.findall(r'\d+' ,rat_rev)
            n_rev=int(rr_list[2]) 
            #number_review.append(n_rev)

    except:
            n_rev= nrev_list[number]
    
    
    if n_rev//10 == n_rev/10:  #se il numero di recensioni è divisibile per 10
            
        for i in range(1,int(n_rev/10 +1),1):  #max 10 reviews per pagina
                
            time.sleep(1)  
            #a inizio pagina viene creata la lista ausiliaria dei link delle reviews, che verranno catturate in un secondo momento
           
            try:
                    driver.get(url+ "&page=" +str(i))
            except:
                    time.sleep(6)
                    driver.get(url+ "&page=" +str(i))
            
            try:
                try:
                    time.sleep(1)
                    #lista delle 10 review
                    li_reviews=driver.find_elements_by_xpath("//ul[@id='reviewList']/li")
                    if len(li_reviews) == 0:
                        creaerrore #crea un errore e si assa all'exceot
                except:
                    time.sleep(5)
                    #lista delle 10 review
                    li_reviews=driver.find_elements_by_xpath("//ul[@id='reviewList']/li")
                    if len(li_reviews) == 0:
                        creaerrore #crea un errore e si assa all'exceot
            except:
                continue  #se la lista continua ad essere vuota dopo 2 tentativi, passiamo oltre
        
    
            #ciclo per metadati
        
            n=0 #ad ogni cambio pagina refresh
            for el in li_reviews:
                
                #cominciamo con il fillare la lista dei titoli, isbn e genere
                titoli.append(title[number])
                isbn_code.append(isbn[number])
                genere.append(gen[number])               
                
                n+=1 #utile per catturare user e titolo. Conta il numero di elementi in lista
    
                #rating/data
                try:
                    time.sleep(1)
                    rat_dat=el.find_elements_by_class_name('txt_inline')
                    if len(rat_dat)==1:
                        ratings.append('no vote')
                        rev_date.append(rat_dat[0].get_attribute('textContent'))  #solo dat
        
                    else:
                        ratings.append(rat_dat[0].get_attribute('textContent')) #rat
                        rev_date.append(rat_dat[1].get_attribute('textContent'))  #dat
                except:
                        ratings.append('error') #rat
                        rev_date.append('error')  #dat
                
                #userlist
                try:
                    time.sleep(1)
                    us_tit=el.find_element_by_id('review_author_'+str(n))
                    spl=us_tit.text.split(' ')
                    userlist.append(spl[3][:-1]) #username
                except:
                    userlist.append('error') #username      
                
                #content
                try:
                    time.sleep(1)
                    revlnk=el.find_element_by_xpath("//dd[@id='review_author_"+str(n)+"']/a").get_attribute('href') 
                    linkrev.append(revlnk)
                   
                    
                except:
                    linkrev.append('no link')
                
                try:
                    linkrev.append(str(spl[0]))  #proviamo a tenerci una seconda via per catturare il link
                    
                except:
                    time.sleep(1)
                    
                    
                   
            
            
            
                
                
    else:  #se il numero di recensioni non è divisibile per 10
            
        for i in range(1,int(n_rev/10 +2),1):  #max 10 reviews per pagina
                
            time.sleep(1)  
            #a inizio pagina viene creata la lista ausiliaria dei link delle reviews, che verranno catturate in un secondo momento
           
            try:
                    driver.get(url+ "&page=" +str(i))
            except:
                    time.sleep(6)
                    driver.get(url+ "&page=" +str(i))
            
            try:
                try:
                    time.sleep(1)
                    #lista delle 10 review
                    li_reviews=driver.find_elements_by_xpath("//ul[@id='reviewList']/li")
                    if len(li_reviews) == 0:
                        creaerrore #crea un errore e si assa all'exceot
                except:
                    time.sleep(5)
                    #lista delle 10 review
                    li_reviews=driver.find_elements_by_xpath("//ul[@id='reviewList']/li")
                    if len(li_reviews) == 0:
                        creaerrore #crea un errore e si assa all'exceot
            except:
                continue  #se la lista continua ad essere vuota dopo 2 tentativi, passiamo oltre
        
    
            #ciclo per metadati
        
            n=0 #ad ogni cambio pagina refresh
            for el in li_reviews:
                
                #cominciamo con il fillare la lista dei titoli, isbn e genere
                titoli.append(title[number])
                isbn_code.append(isbn[number])
                genere.append(gen[number])               
                
                n+=1 #utile per catturare user e titolo. Conta il numero di elementi in lista
    
                #rating/data
                try:
                    time.sleep(1)
                    rat_dat=el.find_elements_by_class_name('txt_inline')
                    if len(rat_dat)==1:
                        ratings.append('no vote')
                        rev_date.append(rat_dat[0].get_attribute('textContent'))  #solo dat
        
                    else:
                        ratings.append(rat_dat[0].get_attribute('textContent')) #rat
                        rev_date.append(rat_dat[1].get_attribute('textContent'))  #dat
                except:
                        ratings.append('error') #rat
                        rev_date.append('error')  #dat
                
                #userlist
                try:
                    time.sleep(1)
                    us_tit=el.find_element_by_id('review_author_'+str(n))
                    spl=us_tit.text.split(' ')
                    userlist.append(spl[3][:-1]) #username
                except:
                    userlist.append('error') #username      
                
                #content
                try:
                    time.sleep(1)
                    revlnk=el.find_element_by_xpath("//dd[@id='review_author_"+str(n)+"']/a").get_attribute('href') 
                    linkrev.append(revlnk)
                   
                    
                except:
                    linkrev.append('no link')
                
                try:
                    linkrev.append(str(spl[0]))  #proviamo a tenerci una seconda via per catturare il link
                    
                except:
                    time.sleep(1)
                    
                    
        
    #droppiamo eventuali duplicati dalla lista dei link
    linkrev2=[]
    for el in linkrev:
            if el not in linkrev2:
                linkrev2.append(el)
                
    #una volta terminate le pagine proviamo a ciclare la lista dei link delle reviews, per estrarne il contenuto
    #content:

    for lnk in linkrev2:
            
            
            
            try:
                time.sleep(1)
                driver.get(str(lnk))
                codice=int(lnk.split("/")[-1])
            except:
                rev_content.append('error')
                continue
            
            try:
                time.sleep(2)
                #entriamo nel frame, tanto è uguale per tutti
                frame = driver.find_element_by_name('mainFrame')
                driver.switch_to.frame(frame)
                #OPERAZIONI PER IL CONTENT
                        
                try:
                            time.sleep(1)
                            
                            try: #blog di tipo1
                                
                                cont=driver.find_element_by_xpath("//div[@class='se_textView']/p[@class='se_textarea']").get_attribute('textContent')
                                rev_content.append(cont)
                                
                            except: #blog tipo2
                                #time.sleep(3)

                                contenuto=driver.find_elements_by_xpath("//div[@class='se-module se-module-text']/p[@class='se-text-paragraph se-text-paragraph-align-center ']")
                                if len(contenuto) == 0:
                                    #devo provocare l'errore per passare al prossimo except
                                    contenuto=driver.find_element_by_xpath("//div[@class='se-module se-module-text']/p[@class='se-text-paragraph se-text-paragraph-align-center ']")
                                else:
                                    str_cont=''
                                    for elem in contenuto:
                                        cont=elem.get_attribute('textContent')
                                        str_cont+=str(cont)+' '
                                    rev_content.append(str_cont)
                                
                except: #blog tipo3
                            
                            try:
                                contenuto=driver.find_elements_by_xpath("//div[@id='post-view"+str(codice)+"']/p").get_attribute('textContent')
                                if len(contenuto) == 0:
                                    #devo provocare l'errore per passare al prossimo except
                                    contenuto=driver.find_element_by_xpath("//div[@id='post-view"+str(codice)+"']/p").get_attribute('textContent')
                                else:
                                    str_cont=''
                                    for elem in contenuto:
                                        cont=elem.get_attribute('textContent')
                                        str_cont+=str(cont)+' '
                                    rev_content.append(str_cont)
                                    
                            except: #in quanto ci sono casi che con la p non funzionano
                                contenuto=driver.find_element_by_xpath("//div[@id='post-view"+str(codice)+"']").get_attribute('textContent')
                                rev_content.append(contenuto)
                                
                                 
                        
                        
            except:
                        rev_content.append('error')
                    
            #numero di likes si predne da qui:
            
            try:
                like=driver.find_element_by_class_name('u_cnt._count').get_attribute('textContent')
                n_likes.append(like)
            except:
                n_likes.append('error')
                
            
                
                
            driver.switch_to.default_content() #passiamo da frame a default content
        
                
                
    #a fine ciclo uniformiamo le lunghezze delle liste. Teniamo la lunghezza delle reviews
    titoli=titoli[:len(rev_content)]
    isbn_code=isbn_code[:len(rev_content)]
    genere=genere[:len(rev_content)]
    ratings=ratings[:len(rev_content)]
    userlist=userlist[:len(rev_content)]
    n_likes=n_likes[:len(rev_content)]
    rev_date=rev_date[:len(rev_content)]
    
    print(number)
    #print('\n')
    print(len(titoli))
    print(len(userlist))
    #print('\n')
    
    #vediamo quante sono vuote
    err=0
    for el in rev_content:
        if el == '' or el == 'error':
            err+=1
    print(err)
    print('\n')
    

    
        

    
        
        
```

```python
#controllo duplicati
naver_reviews2=naver_reviews.drop_duplicates( subset=['Title','ISBN','Username','Review_Content'],keep='first',inplace=False)
naver_reviews2=naver_reviews2.reset_index()
naver_reviews2
```

```python
#selezione lingua corretta
from langdetect import detect
lang=[]
for index, row in naver_reviews2['Review_Content'].iteritems():
    rev=row.replace('\n','').replace('\t','').replace(' ','')
    try:
        language = detect(rev) #detecting each row
        lang.append(language)
    except:
        lang.append('not found')
print(lang[:10])
```

```python
ind=[]
count=0
for el in lang:
    if el != 'ko':
        ind.append(count)
    count+=1
print(ind)
```
