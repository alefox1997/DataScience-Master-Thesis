```python
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
PATH='C:\\Users\\Fossati\\Desktop\\Tesi_Ale\\chromedriver91.exe'
driver = webdriver.Chrome(PATH)
```

```python
meta_yes=pd.read_csv("\\Users\\Fossati\\Desktop\\Tesi_Ale\\Codice\\Codice_KO\\Yes24Meta_KO.csv")
meta_yes
#per koreano tengo tutto in una cartella unica con codici e file, così da avere tutto li anche in vista della pulizia della cartella
```

```python
#LISTA LINK DEFINITIVA
#codice creazione link a partire da ISBN del file di YES24

isbn_list= meta_yes['ISBN'].to_list()
lista_lnk=[]
url='https://book.naver.com/search/search.nhn?query='
ind_libro=-1
for el in isbn_list:
    ind_libro+=1
    lnk=url+str(el)
    driver.get(lnk)
    try:
        libri=driver.find_element_by_xpath("//div[@id='content']/ul[@class='basic']/li/div[@class='thumb type_search']/div[@class='thumb_type thumb_type2']/a").get_attribute('href')
        lista_lnk.append(libri)
    except:
        lista_lnk.append('error')
        print(ind_libro)
print('\n')
print(len(lista_lnk))
print(lista_lnk[:20])
    
```

```python
#273, 607, 612, 631, 906 non hanno link
```

```python
#metadati da catturare

prices=[]
num_revs=[]
overall_rat=[]
pub_date=[]
```

```python
#prima parte in cui si scrapano info generali sul libro di riferimento


numero_libro=-1
no_link=0
# queste  liste saranno unite in un dataframe a unico con ogni riga equivalente ad un libro
for book in lista_lnk:   #togli poi 21
    
    try:
        try:
            url= str(book)
            driver.get(url)
        except:
            time.sleep(5)
            url= str(book)
            driver.get(url)
    except:
        no_link+=1
        prices.append('error')
        num_revs.append('error')
        overall_rat.append('error')
        pub_date.append('error')
        continue
    #prima parte, caratteristiche generali del libro:
    
        
    
    #overall_rating e numero di reviews
    
    try:
        time.sleep(2)
        #rat_rev=driver.find_element_by_id('txt_desc_point').text
        rat_rev=driver.find_element_by_xpath("//div[@class='book_info_inner']/div[@class='txt_desc']/a[@id='txt_desc_point']").text
        rr_list=re.findall(r'\d+' ,rat_rev)
        overall_rat.append(float(str(rr_list[0])+"."+str(rr_list[1])))  #questo lo appendi alla lista rating
        num_revs.append(int(rr_list[2]))  #questo lo appendi alla lista n_reviews per libro
    except:
        overall_rat.append('error')
        num_revs.append('error')
    
    #prezzo
    try:
        time.sleep(1)
        prezzo=driver.find_element_by_class_name('price_area').text
        price=re.findall(r'\d+' ,prezzo)
        prices.append(str(price[0])+"."+str(price[1]))
    except:
        prices.append('error')
    
    
    try:
        try:
            time.sleep(1)
            member=driver.find_elements_by_xpath("//div[@class='book_info_inner']/div")
        except:
            time.sleep(5)
            member=driver.find_elements_by_xpath("//div[@class='book_info_inner']/div")
    except:
        #authors.append('error')
        #isbn.append('error')
        pub_date.append('error')
        #n_pages.append('error')
        continue
    
    #pub date
    try:
        time.sleep(1)
        aut_dat=member[1].text
        aut_dat=aut_dat.split("|")

    #valuta se la lista è da 2, 3,4 o 5 elementi in quanto possono esserci anche il designer, il publisher ad esempio oltre a autore e data

        if len(aut_dat) == 5:
            #authors.append(aut_dat[0].split(" ")[1])  #nome autore
            pub_date.append(aut_dat[4])  #data di pubblicazione
        if len(aut_dat) == 4:
            #authors.append(aut_dat[0].split(" ")[1])  #nome autore
            pub_date.append(aut_dat[3])  #data di pubblicazione
        if len(aut_dat) == 3:
            #authors.append(aut_dat[0].split(" ")[1])  #nome autore
            pub_date.append(aut_dat[2])  #data di pubblicazione
        if len(aut_dat) == 2:
            #authors.append(aut_dat[0].split(" ")[1])  #nome autore
            pub_date.append(aut_dat[1])  #data di pubblicazione, supponendo non ci sia il publisher
            
    except:
        #authors.append('error')
        pub_date.append('error')        

    

        
        
    






```
