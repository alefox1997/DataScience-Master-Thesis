```python
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
PATH='C:\\Users\\aless\\OneDrive\\Desktop\\Laurea_magistrale\\Tesi\\chromedriver.exe'
driver = webdriver.Chrome(PATH)
```

```python
#importo il file di AMAZON IT ed estraggo la lista degli ISBN (o dei titoli se con ISBN non riesci)

amazon_ita=pd.read_csv("C:\\Users\\aless\\OneDrive\\Desktop\\Laurea_magistrale\\Tesi\\Dati\\Dati_Italia\\Meta_Amazon_IT.csv")
amazon_ita=amazon_ita.drop('Unnamed: 0',axis=1)
amazon_ita
```

```python
#creiamo una lista di link anobii che hanno tutti lo stesso formato

isbn=amazon_ita['ISBN'].tolist()
link='https://www.anobii.com/search//books'
anobii_lnk=[]
for el in isbn:
    anobii = link[:30] + el + link[30:]
    anobii_lnk.append(anobii)
print(len(anobii_lnk))
anobii_lnk[:20]
```

```python
#prove per generi
driver.get('https://www.anobii.com/books/viola-e-il-blu/9788831004114/01b0d482bc8f32c2f1')
g=driver.find_element_by_xpath("//app-genres-edition['_ngcontent-serverapp-c117 _nghost-serverapp-c191']/div[@class='row col']/a")
#for el in g:
    #print(el.get_attribute('textContent'))
g.get_attribute('textContent')
```

```python
num_pag=[]
stelle=[]
num_voti=[]
num_rece=[]
#in aggiunta:
pub_date=[]
lingua=[]
sug_gen=[]
```

```python
no_lnk=0
for lnk in anobii_lnk:

    time.sleep(2)
    count=0
    #otteniamo il link con ISBN
    try:
        try:
            driver.get(lnk)
        except:
            time.sleep(3)
            driver.get(lnk)
    except:
        num_pag.append('no link')
        num_voti.append('no link')
        num_rece.append('no link')
        stelle.append('no link')
        pub_date.append('no link')
        lingua.append('no link')
        sug_gen.append('no link')
        no_lnk+=1
        
    
    try:
        try:
            lnk2=driver.find_element_by_xpath("//div[@class='col-cover']/a['_ngcontent-serverapp-c194']").get_attribute('href')
            driver.get(lnk2)
            count+=1   #flag per attivare codice
        except:
            time.sleep(3)
            lnk2=driver.find_element_by_xpath("//div[@class='col-cover']/a['_ngcontent-serverapp-c194']").get_attribute('href')
            driver.get(lnk2)
            count+=1   #flag per attivare codice            
        
    except:
        num_pag.append('no link')
        num_voti.append('no link')
        num_rece.append('no link')
        stelle.append('no link')
        pub_date.append('no link')
        lingua.append('no link')
        sug_gen.append('no link')
        no_lnk+=1
        
    
    if count==1:
    #ora che siamo dentro la pagina delle statistiche:
    #numero pagine:
        try:
            pag=driver.find_elements_by_xpath("//app-about-edition['_ngcontent-serverapp-c117']['_nghost-serverapp-c186']/div[@class='ng-star-inserted']/div['_ngcontent-serverapp-c186']")
            n_pag=pag[5].text
            n_pag=int(re.findall(r'\d+' ,n_pag)[0])
            num_pag.append(n_pag)
            time.sleep(1)
        except:
            num_pag.append('error')
            
    #pubdate 
        try:
            #pag=driver.find_elements_by_xpath("//app-about-edition['_ngcontent-serverapp-c117']['_nghost-serverapp-c186']/div[@class='ng-star-inserted']/div['_ngcontent-serverapp-c186']")
            date=pag[2].text
            pub_date.append(date)
            time.sleep(1)
        except:
            pub_date.append('error')
    
    #lingua
        
        try:
            #pag=driver.find_elements_by_xpath("//app-about-edition['_ngcontent-serverapp-c117']['_nghost-serverapp-c186']/div[@class='ng-star-inserted']/div['_ngcontent-serverapp-c186']")
            lang=pag[0].text
            lingua.append(lang)
            time.sleep(1)
        except:
            lingua.append('error')
            
    #genere suggerito:
        aus_gen=[]
        try:
            g=driver.find_elements_by_xpath("//app-genres-edition['_ngcontent-serverapp-c117 _nghost-serverapp-c191']/div[@class='row col']/a")
            if len(g)!=0: #ci sono + generi
                for el in g:
                    aus_gen.append(el.get_attribute('textContent'))
                sug_gen.append(aus_gen)
            
            else: #sarà solo 1 genere
                g2=driver.find_element_by_xpath("//app-genres-edition['_ngcontent-serverapp-c117 _nghost-serverapp-c191']/div[@class='row col']/a")
                sug_gen.append(g2.get_attribute('textContent'))
        except:#non ci sarà genere
            sug_gen.append('error')
    
    #numero voti
        try:
            time.sleep(3)
            n_stars=driver.find_element_by_xpath("//div[@class='expandible']/span[@class='star-count']/span[@class='color-yellow']").text
            try:
                n_stars=n_stars.replace('.','')
                num_voti.append(int(n_stars))
            except:
                n_stars=n_stars.replace(',','')
                num_voti.append(int(n_stars))                
            #time.sleep(1)
        except:
            num_voti.append('error')
    
    #numero recensioni
        try:
            try:
                time.sleep(1)
                n_rev=driver.find_element_by_xpath("//app-number-text['_ngcontent-serverapp-c117']/a[@class='ng-star-inserted']").text
                try:
                    n_rev=n_rev.replace('.','')
                    num_rece.append(int(n_rev))
                except:
                    n_rev=n_rev.replace(',','')
                    num_rece.append(int(n_rev))                   
                #time.sleep(1)
            except:
                n_rev=driver.find_element_by_xpath("//app-number-text['_ngcontent-serverapp-c117']/span[@class='no-anchor ng-star-inserted']").text
                try:
                    n_rev=n_rev.replace('.','')
                    num_rece.append(int(n_rev))
                except:
                    n_rev=n_rev.replace(',','')
                    num_rece.append(int(n_rev))
        except:
            num_rece.append('error')
            
    
    #stelle
        try:
            stars=driver.find_element_by_xpath("//div[@class='expandible']/ngb-rating['_ngcontent-serverapp-c178']").get_attribute('aria-valuenow')
            stelle.append(round(float(stars),2))
        #time.sleep(1)
        except:
            stelle.append('error')
    
    #teniamo conto errori ad ogni ciclo
    print(no_lnk)
    print(len(num_rece))
    er=0
    for el in num_rece:
        if el=='error':
            er+=1
    print(er)
    print('\n')
    


```
