```python
def avg_rev_book (data1, col1, col2): #passiamo il dataset e le due colonne per cui computare media e intervallo di confidenza o scarto
   # vogliamo ottenere dati per retail e non retail del dataset che gli passiamo
    ret=data1[col1].tolist()
    ret2=[]
    for num in ret:
        try:
            int(float(num) ) #si filtrano i nan
            ret2.append(int(float(num) ))
        except:
            continue
    ret2=pd.Series(ret2)
    avg_ret=float(round(ret2.sum()/len(ret2),2))
    std_ret=np.std(ret2)
    conf_ret=confint(avg_ret, std_ret, len(ret2))
    scarto_ret=float(round((conf_ret[1]-conf_ret[0])/2,2))
    print(avg_ret)
    print(scarto_ret)
    print('\n')
    
    noret=data1[col2].tolist()
    noret2=[]
    for num in noret:
        try:
            int(float(num) ) #si filtrano i nan
            noret2.append(int(float(num) ))
        except:
            continue
    noret2=pd.Series(noret2)
    avg_noret=float(round(noret2.sum()/len(noret2),2))
    std_noret=np.std(noret2)
    conf_noret=confint(avg_noret, std_noret, len(noret2))
    scarto_noret=float(round((conf_noret[1]-conf_noret[0])/2,2))
    print(avg_noret)
    print(scarto_noret)
    print('\n')
    
    return 'ok'
```

```python
#funzione per calcoare intervalli di confidenza per normale
import math
def confint(media, stdev, n):
    lower=media-1.64*stdev/(math.sqrt(n))
    upper=media+1.64*stdev/(math.sqrt(n))
    scarto = 1.64*stdev/(math.sqrt(n))
    intconf=(round(lower,2),round(upper,2))
    return intconf
```

```python
def distr_rat(dataset, colname, lang):
    rat_clean=[]
    for el in dataset[colname].tolist():
        try:
            rat_clean.append(int(float(el))) #con la logica che il float si converte in int, tutto il resto è scartato
        except:
            continue
    #print(rat_clean)
        
    values=pd.Series(rat_clean).to_frame().groupby(0).size()#.tolist()
    #print(values)
    perc=[]
    for el in values:
        perc.append(round(float(el*100/len(rat_clean))))
        
    if lang=='ko':
        #print(perc)
        perc2=[]
        for ele in range(0, len(perc), 2):
            perc2.append(perc[ele] + perc[ele + 1])
        perc=perc2
            
    return perc  #ci ritorna la lista di percentuali , da 1 stella a 5 stelle
```

```python
import matplotlib.pyplot as plt
import numpy as np
def pie(lista, retail, loca):
    y = np.array(lista)
    mylabels = ["1 Star","2 Stars", "3 Stars", "4 Stars", "5 Stars"]
    percent=[]
    #for el in y:
        #percent.append(str(el)+"%")
    if retail == 'yes':
        myexplode = [0, 0, 0, 0, 0.2]
    else:
        myexplode = [0, 0, 0, 0, 0]
    plt.pie(y, explode = myexplode, shadow = True,autopct='%.2f%%')#labels= percent
    plt.legend(mylabels,loc=loca)
    return plt.show() 
```

```python
#test lunghezza media delle recensioni per lingua non può riguardare il numero di caratteri o parole per frase, in quanto
#la struttura dei linguaggi occidentali e coreano è totalmente differente, per cui ci limiteremo a rispondere alle RQ:

#quale è il numero medio di frasi contenuto in una recensione del nostro campione per ogni lingua e per diversa piattaforme per shared ed in generale?
# due barchart semplici con IC, come quelli per rating


#numero medio di frasi per genere per lingua su libri geenrali, perchè per gli altri non abbiamo genere pronto?
#anche qui barchart, vedi se riesci a variare
```

```python
def avg_length(dataset,colname,lang, n):
    rev=dataset[colname].tolist()
    totfrasi=[]
    
    if lang=='ko':
        for review in rev:
            perpunto=review.replace("!","!.").replace("?","?.").replace("|",".").split(". ") #per ko spazio punto
            totfrasi.append(len(perpunto))
    else:
        for review in rev:
            perpunto=review.replace("!","!.").replace("?","?.").replace("|",".").split(".") #en o it solo punto
            totfrasi.append(len(perpunto))            
        

        
    avg_frasi_per_rev=float(round(sum(totfrasi)/len(totfrasi),2))
    std=float(round(np.std(totfrasi),2))
    int_conf=confint(avg_frasi_per_rev, std, n)  #no problema sulla normalità, dato che abbiamo tanti dati TCL
        
    return avg_frasi_per_rev, std, int_conf  
```

```python
#funzione per calcoare intervalli di confidenza per normale

def confint(media, stdev, n):
    lower=media-1.96*stdev/(math.sqrt(n))
    upper=media+1.96*stdev/(math.sqrt(n))
    scarto = 1.96*stdev/(math.sqrt(n))
    intconf=(round(lower,2),round(upper,2))
    return intconf
```

```python
import numpy as np
import math

def mean_sent(dataset):

    score1=dataset['Sent_Score'].tolist()
    score = []
    for el in score1:
        try:
            score.append(float(el))
        except:
            continue 
    media=round(np.mean(score),2)
    stdev= round(np.std(score),2)
    n=len(dataset)
    conf=confint(media, stdev, n)
    scarto=round((conf[1] -conf[0])/2,2)
    return media, scarto
```

```python
import matplotlib.pyplot as plt
import numpy as np
def pie(lista,titolo):
        y = np.array(lista)
        mylabels = ['Anger', 'Fear', 'Joy', 'Neutral', 'Sadness', 'Surprise']
        plt.pie(y, shadow = True,autopct='%.2f%%')#labels= percent
        plt.legend(mylabels,loc="upper right",bbox_to_anchor=(1.2,1.025))
        plt.title(titolo)
        return plt.show()
    


def emotion_pie(dataset, titolo):
    import ast
    new_emo=[] #lista di tutte le emozioni
    for index,row in dataset.iterrows():
      try:
        emo=ast.literal_eval(row['Emotions'])
        if 'Neutral' in emo:
            if len(emo) != 1:
                emo.remove('Neutral')
        for emotion in emo:
            new_emo.append(emotion)
      except:
        new_emo.append(row['Emotions']) #per naver, che presenta alcuni neutral fuori dalla lista
            
    new_emo.sort()  #sorting della lista per ordine alfabetico Anger, Fear, Joy, Neutral, Sadness, Surprise
    df=pd.DataFrame({'Emo':new_emo})
    lista=list(df['Emo'].value_counts(sort=False).sort_index(ascending=True))
    
    piechart_emo=pie(lista, titolo)
    
 
    
    return piechart_emo
```

```python
#funzione generica che potrebbe tornare utile

def prop_test(p_a, n_a, p_b, n_b, hp):

    from statsmodels.stats.proportion import proportions_ztest
    import numpy as np

    significance = 0.025

    sample_success_a, sample_size_a = (p_a, n_a)
    sample_success_b, sample_size_b = (p_b, n_b)

    successes = np.array([sample_success_a, sample_success_b])
    samples = np.array([sample_size_a, sample_size_b])

    stat, p_value = proportions_ztest(count=successes, nobs=samples,  alternative=hp) #scegli il tipo di test

    print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

    if p_value > significance:
       print ("Fail to reject the null hypothesis - we have nothing else to say")
    else:
       print ("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

```python

#funzione specifica

from statsmodels.stats.proportion import proportions_ztest
import numpy as np



#https://sonalake.com/latest/hypothesis-testing-of-proportion-based-samples/

def two_prop_test(dataset1, dataset2, col1, col2, lang): #1 per piattaforma commerciale, 2 per piattaforma no retail
# can we assume anything from our sample
    significance = 0.025
    
    #dobbiamo trasformare tutti i dati delle colonne interessate in float
    lista1=[]
    lista2=[]
    
    l1=dataset1[col1].tolist()
    l2=dataset2[col2].tolist()
    
    count_comm=0
    count_gen=0
    
    if lang != 'ko':

        for el in l1:
            if el != 'nan':
                try:
                    var=float(int(float(el)))
                    lista1.append(var)
                    if var == 1.0 or var == 5.0:
                        count_comm+=1
                    
                    
                except:
                    continue
                    
                    
                    
            
        for el in l2:
            if el != 'nan':
                try: #escludo errors
                    var=float(int(el))
                    lista2.append(var) #per escludere nan
                    if var == 1.0 or var == 5.0:
                        count_gen+=1
                except:
                    continue
                
    
    
    
    else: #per gestire la faccenda di yes24
        
        
        for el in l1:
            str_tolist = re.findall(r'\d+', el)  #rende la stringa una lista, che può avere lunghezza 0,1 o 2
    
            if len(str_tolist) == 2:
                var = float((float(str_tolist[0]) + float(str_tolist[1]))/2)
                lista1.append(var)
                if var == 1.0 or var == 5.0:
                        count_comm+=1
            elif len(str_tolist) == 1:
                var=float(float(str_tolist[0]))
                lista1.append(var)
                if var == 1.0 or var == 5.0:
                        count_comm+=1
    
            else:
                continue #caso di error o 

                        
                        
        for el in l2:
            try: #escludo errors
                var=float(el)/2
                lista2.append(var)  #naver punteggio in decimi
                if var == 1.0 or var == 5.0:
                    count_gen+=1
            except:
                continue
                
                
                
        #ora abbiamo normalizzato le liste di scores
        
        

    sample_success_a, sample_size_a = (count_comm, len(lista1))
    sample_success_b, sample_size_b = (count_gen, len(lista2))


    successes = np.array([sample_success_a, sample_success_b])
    samples = np.array([sample_size_a, sample_size_b])


    stat, p_value = proportions_ztest(count=successes, nobs=samples,  alternative='larger')

# report
    print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

    if p_value > significance:
       print ("Fail to reject the null hypothesis - we have nothing else to say")
    else:
       print ("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np



#https://sonalake.com/latest/hypothesis-testing-of-proportion-based-samples/

def two_prop_test_rating(dataset1, dataset2, col1, col2): 
# can we assume anything from our sample
    significance = 0.01
    
    polarized=[0.5, '0.5', 1, 1.0, '1', '1.0',5, 5.0, '5', '5.0']
    
    lista1=[]
    lista2=[]
    
    l1=dataset1[col1].tolist()
    l2=dataset2[col2].tolist()
    
    count_comm=0
    count_gen=0
    #si ciclano separati per tenere conto della numerosità campionaria diversa
    for el in l1:
        if el != 'error':
                try:
                    lista1.append(el)  #evita nan
                    if el in polarized:
                        count_comm+=1
                except:
                    continue
            
    for el in l2:
        if el != 'error':
                try:
                    lista2.append(el)  
                    if el in polarized:
                        count_gen+=1
                except:
                    continue
        
        

    sample_success_a, sample_size_a = (count_comm, len(lista1))
    sample_success_b, sample_size_b = (count_gen, len(lista2))


    successes = np.array([sample_success_a, sample_success_b])
    samples = np.array([sample_size_a, sample_size_b])


    stat, p_value = proportions_ztest(count=successes, nobs=samples,  alternative='larger')

# report
    print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

    if p_value > significance:
       print ("Fail to reject the null hypothesis - we have nothing else to say")
    else:
       print ("Reject the null hypothesis - suggest the alternative hypothesis is true")
    
```
