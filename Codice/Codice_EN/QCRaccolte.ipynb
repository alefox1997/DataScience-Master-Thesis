```python
tit_date=goodreads_def[['Titolo','Rev_Date_GR']]
tit_date
```

```python
#filtro di valori che hanno piu di 300 reviews a ausa delll'errore di data. Prendiamo
ind=0
und=[]

for index,row in tit_date.iterrows():
    #vediamo se il titolo √® presente pi√π di 300 volte
    titolo = row['Titolo']
    lung = len(tit_date[tit_date['Titolo'] == titolo])
    if lung > 300:
        anno = row['Rev_Date_GR'][-4:]
        if int(anno)<2020:
            und.append(ind)
        
        
    ind+=1 
len(und)
```

```python
#primo filtro per date minori di 2020 per libri che hanno piu di 300 rev
goodreads_filter=goodreads_def.drop(und,axis=0).reset_index()
goodreads_filter
```

```python
prova=goodreads_filter.groupby(['Titolo'],sort=False).size().reset_index()
prova

```

```python
ind=[] #lista indici da droppare
n_rev=-1
for index,row in prova.iterrows(): #303
    numero=row[0]
    n_rev+=numero
    if numero > 300:
        n_ind=numero-300
        x=0
        while x < n_ind:
            ind.append(n_rev-x) #appendi gli indici al contrario, poi ordinerai la lista
            x+=1
            
print(len(ind))
ind[:10]
```

```python
#primo filtro per date minori di 2020 per libri che hanno piu di 300 rev
goodreads_filter2=goodreads_filter.drop(ind,axis=0).drop('level_0',axis=1).reset_index()
goodreads_filter2
```

```python
prova2=goodreads_filter2.groupby(['Titolo']).size().reset_index()
prova2[prova2[0]>300] #sistemato

```
