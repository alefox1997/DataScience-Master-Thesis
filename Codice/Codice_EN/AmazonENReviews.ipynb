```python
#lista numero reviews amazon
nrev_list=meta_en['N_Reviews_AZ'].to_list()
nrev_list[:10]
```

```python
#creiamo liste vuote per poter poi fare i dataframe
#per queste due liste, ogni volta che passiamo in rassegna ad un oggetto webdriver il li_reviews, appendiamo anche a loro
titoli=[]
isbn_code=[]
genere=[]

#metadata reviews
userlist=[]
ratings=[]
rev_date=[]
rev_place=[]
rev_content=[]
n_likes=[]
```

```python
number= -1 #-1
#numero_review=0  #per contare il numero di reviews, al fine di vedere se si ottiene un campione rappresentativo o meno numericamente parlando

for link in mylinks: #limita per le prove
    number+=1 #per tenere conto del libro a cui siamo arrivati in caso di stop e per riempire le liste isbn e titolo
 
    flag=0
    #time.sleep(1)
    try:
        try:
            driver.get(link)
        except:
            time.sleep(5)
            driver.get(link)
    except:

        userlist.append('no_link')
        ratings.append('no_link')
        rev_date.append('no_link')
        rev_content.append('no_link')
        n_likes.append('error')
        rev_place.append('error')
        titoli.append(title[number])
        isbn_code.append(isbn[number])
        genere.append(gen[number])
        continue
        
        
    
    #codice per riuscire a prendere il libro di interesse. capita che ci siano molte edizioni e in testa alla pagina amazon ci sia quella meno recensita, non originale
    #poi avrai i link quindi potrai rimuoverlo
    try:
        try:
            n_rev_aus=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-micro']/div[@class='a-row a-size-small']/span['aria-label']/a[@class='a-link-normal']/span")
        except:
            time.sleep(5)
            n_rev_aus=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-micro']/div[@class='a-row a-size-small']/span['aria-label']/a[@class='a-link-normal']/span")        
    except:

        userlist.append('no_link')
        ratings.append('no_link')
        rev_date.append('no_link')
        rev_content.append('no_link')
        n_likes.append('error')
        rev_place.append('error')
        titoli.append(title[number])
        isbn_code.append(isbn[number])
        genere.append(gen[number])
        continue
    
    c=0
    aus=[]
    try:
        for el in n_rev_aus[:5]:
            rep=n_rev_aus[c].text.replace('.','')
            aus.append(int(rep))
            c+=1
        max_value = max(aus)
        max_index = aus.index(max_value)
    
    except:
        try:
            for el in n_rev_aus[:3]:
                rep=n_rev_aus[c].text.replace('.','')
                aus.append(int(rep))
                c+=1
            max_value = max(aus)
            max_index = aus.index(max_value)
        except:
        
            max_index=0

    try:
        try:
            n_rev_lnk=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-micro']/div[@class='a-row a-size-small']/span['aria-label']/a")
        except:
            time.sleep(5)
            n_rev_lnk=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-micro']/div[@class='a-row a-size-small']/span['aria-label']/a")
    except:

        userlist.append('no_link')
        ratings.append('no_link')
        rev_date.append('no_link')
        rev_content.append('no_link')
        n_likes.append('error')
        rev_place.append('error')
        titoli.append(title[number])
        isbn_code.append(isbn[number])
        genere.append(gen[number])
        continue
    
    try:
        url=str(n_rev_lnk[max_index].get_attribute('href'))
        driver.get(url) #si acquisice l'url per lo scraping a sto punto
        flag+=1
    except:

        userlist.append('no_link')
        ratings.append('no_link')
        rev_date.append('no_link')
        rev_content.append('no_link')
        n_likes.append('error')
        rev_place.append('error')
        titoli.append(title[number])
        isbn_code.append(isbn[number])
        genere.append(gen[number])

    
    if flag==1:

        #numero recensioni
        try:
            try:
                time.sleep(1)
                all_rev=driver.find_element_by_xpath("//a[@data-hook='see-all-reviews-link-foot']")
                n_rev_link=all_rev.get_attribute('href')
                url2=str(n_rev_link+str('&sortBy=recent'))
                driver.get(url2)
                #n_rev=driver.find_element_by_xpath("//div[@data-hook='cr-filter-info-review-rating-count']/span")
                #v_r=n_rev.text.split('|')  #per valutare se sono sopra le 300 o meno
                #n_rev=int(re.findall(r'\d+' ,v_r[1].replace('.',''))[0])
                #numero_review+=n_rev
            except:
                time.sleep(5)
                driver.get(url)
                all_rev=driver.find_element_by_xpath("//a[@data-hook='see-all-reviews-link-foot']")
                n_rev_link=all_rev.get_attribute('href')
                url2=str(n_rev_link+str('&sortBy=recent'))
                driver.get(url2)
                #n_rev=driver.find_element_by_xpath("//div[@data-hook='cr-filter-info-review-rating-count']/span")
                #v_r=n_rev.text.split('|')
                #n_rev=int(re.findall(r'\d+' ,v_r[1].replace('.',''))[0])
            #la logica è sempre quella per cui se dà errore è perchè è sovraccaricato di lavoro
        except:

            userlist.append('no_link')
            ratings.append('no_link')
            rev_date.append('no_link')
            rev_content.append('no_link')
            n_likes.append('error')
            rev_place.append('error')
            titoli.append(title[number])
            isbn_code.append(isbn[number])
            genere.append(gen[number])
            continue
    

            
        for i in range(1,50,1):  # 10 reviews per pagina, alla 50 esima pagina stop, obiettivo 450 reviews (filtrando quelle in italiano)
            
                bandiera=0
                time.sleep(3)
                #metadati recensioni:
                try:
                    try:
                        li_reviews=driver.find_elements_by_xpath("//div[@id='cm_cr-review_list']/div[@data-hook='review']")
                        if len(li_reviews)==0:
                            time.sleep(4)
                            li_reviews=driver.find_elements_by_xpath("//div[@id='cm_cr-review_list']/div[@data-hook='review']")
                        bandiera+=1
    
                    except:
                        time.sleep(5)
                        li_reviews=driver.find_elements_by_xpath("//div[@id='cm_cr-review_list']/div[@data-hook='review']")
                except:
                    userlist.extend(['error' for x in range(10)])
                    ratings.extend(['error' for x in range(10)])
                    rev_date.extend(['error' for x in range(10)])
                    rev_place.extend(['error' for x in range(10)])
                    n_likes.extend(['error' for x in range(10)])
                    rev_content.extend(['error' for x in range(10)])
                    titoli.extend([title[number] for x in range(10)])
                    isbn_code.extend([isbn[number] for x in range(10)])
                    genere.extend([gen[number] for x in range(10)])
                    
                if bandiera==1:

                    for el in li_reviews:
                        
                        time.sleep(1)
                        titoli.append(title[number])
                        isbn_code.append(isbn[number])
                        genere.append(gen[number])
                    
                    #username
                    
                        try:
                            user=el.find_element_by_class_name('a-profile-name').text
                            userlist.append(user)
                        
                        except:
                            userlist.append('error')
                      
    
                    #rating review
        
                        try:
                            ids=el.get_attribute('id')
                        except:
                            time.sleep(5)
                            ids=el.get_attribute('id')                    
                    
                        try:              
                            rat=el.find_element_by_xpath("//div[@id='customer_review_foreign-"+str(ids)+"']/div[@class='a-row a-spacing-none']/i[@data-hook='cmps-review-star-rating']/span").get_attribute('textContent')
                            rat=str(rat).replace(',','.')
                            ratings.append(float(rat[0:3])) #sono sempre le prime 3 tanto
                        
                        except:
                            ratings.append('error')
                            
                        
    
                    #data e luogo review
        

                        try:
                        #ids=el.get_attribute('id')
                            rev_dt=el.find_element_by_xpath("//div[@id='"+str(ids)+"-review-card']/div[@class='a-section celwidget']/span[@data-hook='review-date']").text
                            lstring=rev_dt.split(' ')
                            dat=str(str(lstring[-3])+' '+str(lstring[-2])+' '+str(lstring[-1]))
                            rev_date.append(dat)
                            if len(lstring)== 7:
                                pae=str(lstring[2])
                                rev_place.append(pae)
                            if len(lstring)== 8:
                                pae=str(str(lstring[2])+' '+str(lstring[3]))
                                rev_place.append(pae)
                            if len(lstring)== 9:
                                pae=str(str(lstring[2])+' '+str(lstring[3])+' '+str(lstring[4]))
                                rev_place.append(pae)
                       
                        except:
                            rev_date.append('error')
                            rev_place.append('error')
                            
                   
                        #numero di like
                        try:
                            n_lik=driver.find_element_by_xpath("//div[@class='a-row review-comments comments-for"+str(ids)+"']/div/span/div/span").text
                            n_likes.append(n_lik)
                        except:
                            n_likes.append('error')

                            
                            
    
                    
        
                     
                    
                    #content review
                        time.sleep(1)
                        try:
                            try:
                                rev_cont=el.find_element_by_xpath("//div[@id='customer_review_foreign-"+str(ids)+"']/div[@class='a-row a-spacing-small review-data']/span[@data-hook='review-body']/span")
                                rev_content.append(rev_cont.text)
                            except:
                                time.sleep(2)
                                rev_cont=el.find_element_by_xpath("//div[@id='customer_review_foreign-"+str(ids)+"']/div[@class='a-row a-spacing-small review-data']/span[@data-hook='review-body']/span")
                                rev_content.append(rev_cont.text)

                        except:
                            rev_content.append('error')
                            
                            
                            
        

                try:
                    try:
                        succ=driver.find_element_by_xpath("//li[@class='a-last']/a").get_attribute('href')
                        driver.get(str(succ))
                    except:
                        time.sleep(5)
                        succ=driver.find_element_by_xpath("//li[@class='a-last']/a").get_attribute('href')
                        driver.get(str(succ))
                except:
                    if i==49:
                        time.sleep(2)
                    else:
                        break  #significa che è l'ultima pagina per il libro che ha meno di 300 reviews
                    
                    
                    
                    
       
            
    print(number)
    print(len(rev_content))
    print(len(titoli))
    er=0
    for el in rev_content:
        if el=='error':
            er+=1
    print(er)
    print('\n')
    
    if len(rev_content) != len(titoli):   #per capire se ci sono problemi legati al raccoglimento dei ratings
        break
    
```
