```python
import ast

category=[]
media=[]
tag=[]

for index,row in meta_ko.iterrows():
    shelves=meta_ko['Shelves_Yes'][index]
    try:
        shelves=shelves.apply(ast.literal_eval)
        cat=shelves[0]
        cat2=cat.split('\n')
        category.append(cat2[1:])  #perchè il primo è sempre category
    except:
        category.append('error')
        
    try:
    
      if len(shelves) > 1:
    
    
        if shelves[1].split('\n')[0] == '수상내역 및 미디어 추천 분류':
            med=shelves[1]
            media.append(med[1:])
            tag.append('error')

            
        elif shelves[1].split('\n')[0] == '태그 분류':
            ta=shelves[1]
            tag.append(ta[1:])
            media.append('error')
            
        else:
            tag.append('error')
            media.append('error')
    
      else:
        tag.append('error')
        media.append('error')
        
    except:
        tag.append('error')
        media.append('error')
        

            
        
print(len(category))
print(len(media))
print(len(tag))    

#ora che abbiamo risytrutturato la colonna shelves integriamo il file di metadati e lo risalviamo

meta_ko['Shelves_Category']=category
meta_ko['Shelves_Media']=media
meta_ko['Shelves_Tags']=tag
meta_ko2=meta_ko.drop('Shelves_Yes',axis=1)
meta_ko2

```
