```python
yes_link=[]
tipe=[]
aut=[]
lnk='http://www.yes24.com/searchcorner/Search?keywordAd=&keyword=&domain=BOOK&qdomain=%b1%b9%b3%bb%b5%b5%bc%ad&query=&PageNumber=&scode=012'
for autore in aut_list:
    time.sleep(3)

    for i in range(1,30,1):
        #time.sleep(3)
        link= lnk[:112] + str(autore) + lnk[112:124] + str(i) + lnk[124:]
        #print(link)
        try: #perchè può essere che sfori di pagina
            time.sleep(2)
            driver.get(link)
            driver.find_element_by_xpath("//td[@class='goods_infogrp']/p[@class='goods_name goods_icon']").text #è la prova per scorrere le pagine in cui non ci sono libri
        except:
            try: #perchè può essere che sfori di pagina
                #time.sleep(5)
                driver.get(link)
                time.sleep(2)
                driver.find_element_by_xpath("//div[@class='col-cover']/a['_ngcontent-serverapp-c194']")
            except:
            
                break
            
        l=driver.find_elements_by_xpath("//tr/td[@class='goods_img']/a") #i link ai libri yes
        for el in l:
            yes_link.append(el.get_attribute('href'))
            aut.append(autore)


        ti=driver.find_elements_by_xpath("//td[@class='goods_infogrp']/p[@class='goods_name goods_icon']")
        for el in ti:
    
            tipe.append(el.text.split(']')[0][1:])

    
            
y_links=pd.DataFrame(list(zip(yes_link,tipe,aut)),columns=['Link_Yes24','Book_Type','Author_Sugg'])
y_links
```

```python
#quanti non sono books? (ma sono ad esempio audiolibri o cd o ebook?)
indici=[]
count=0
for el in ti:
    if el != '도서':
        indici.append(count)
print(len(indici))
```

```python
#prima parte in cui si scrapano info generali sul libro di riferimento

titles=[]
authors=[]
isbn=[]
yes_link2=y_links['Link_Yes24'].tolist()
```

```python


# queste  liste saranno unite in un dataframe a unico con ogni riga equivalente ad un libro





#scraping metadati

#codice scraping finale yes24


numero_libro=-1   
no_link=0

for book in yes_link2: 
    time.sleep(1)
    numero_libro+=1
    try:
        try:
            url= str(book)
            driver.get(url)
        except:
            time.sleep(5)
            url= str(book)
            driver.get(url)
    except:
        no_link+=1
        continue #non lo registro nemmeno se non trovo il link. Passo oltre
        
        
    
    #titolo
    try:
        title = driver.find_element_by_class_name('gd_name')
        titles.append(title.text)
    except:
        titles.append('error')
    
    #autore
    try:
        author = driver.find_element_by_class_name('lnk_author')
        authors.append(author.text)
    except:
        authors.append('error')
    

    
    
    #numero pagine, isbn e data di pubblicazione:
    try:
        n_pag=driver.find_elements_by_xpath("//div[@id='infoset_specific']/div[@class='infoSetCont_wrap']/div[@class='yesTb']/table[@class='tb_nor tb_vertical']/tbody[@class='b_size']/tr")
    
    except:
        isbn.append('error')

    #isbn
    try:
        isbn.append(re.findall(r'\d+' ,n_pag[2].text)[1])
    except:
        isbn.append('error')
        
    
    

    
    

    
    
  
print(len(titles))
print(len(authors))

print(len(isbn))
```

```python
#LISTA LINK DEFINITIVA
#codice creazione link a partire da ISBN del file di YES24

isbn_list= yes24['ISBN'].to_list()
tit_list= yes24['Title'].to_list()
lista_lnk_naver=[]
url='https://book.naver.com/search/search.nhn?query='
ind_libro=-1
x=0
while x < len(isbn_list):
    time.sleep(1)
    ind_libro+=1
    if isbn_list[x] == 'error' or len(isbn_list[x]) != 13 or isbn_list[x][0] == '\n':
        lnk=url+str(tit_list[x])
    else:
        lnk=url+str(isbn_list[x])
        
    driver.get(lnk)
    try:
        libri=driver.find_element_by_xpath("//div[@id='content']/ul[@class='basic']/li/div[@class='thumb type_search']/div[@class='thumb_type thumb_type2']/a").get_attribute('href')
        lista_lnk_naver.append(libri) #si prende il primo
    except:
        lista_lnk_naver.append('error')
        print(ind_libro)
        
    x+=1
        
print('\n')

    
```
