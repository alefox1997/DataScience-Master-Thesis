```python
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
PATH='C:\\Users\\Fossati\\Desktop\\Tesi_Ale\\chromedriver91.exe'
driver = webdriver.Chrome(PATH)
```

```python
#prima di tutto creiamo la lista contente tutti i link utili per lo scraping
```

```python
#importo il file di AMAZON IT ed estraggo la lista degli ISBN (o dei titoli se con ISBN non riesci)

sh_ita=pd.read_csv("\\Users\\Fossati\\Desktop\\Tesi_Ale\\Dati\\Dati_Italia\\sh_ita_anobii.csv")
#sh_ita=amazon_ita.drop('Unnamed: 0',axis=1)
sh_ita
```

```python
title=[]
author=[]
isbn_code=[]
n_pages=[]
avg_rat=[]
distr_rating=[]
n_voti=[]
n_review=[]
pub_date=[]
lingua=[]
```

```python


#genere suggerito lo prendiamo da anobii per disambiguare


lista_lnk2=sh_ita['Link_Amazon2'].to_list()
for link in lista_lnk2:
    time.sleep(3)
    try:
        try:
            driver.get(link)
        except:
            time.sleep(15)
            driver.get(link)
    except:
        title.append('error')
        author.append('error')
        isbn_code.append('error')
        n_pages.append('error')
        avg_rat.append('error')
        distr_rating.append('error')
        n_voti.append('error')
        n_review.append('error')
        pub_date.append('error')
        
    
    time.sleep(3)
    #titolo
    try:
        time.sleep(3)
        titolo = driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/h1[@id='title']/span[@id='productTitle']")
        title.append(titolo.text)
    except: #potrebbbe esserci il sistema di controllo scraping. Poi re scrapi quelli che non hanno titolo
        try:
            time.sleep(30) #la logica è, se non riesci a prendere il titolo vuol dire ch c'è il captcha, quindi ricarichi la pagina
            driver.get(link)
            titolo = driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/h1[@id='title']/span[@id='productTitle']")
            title.append(titolo.text)
        except:
            time.sleep(1)
            title.append('error')
    
    #autore
    try:
        autore = driver.find_element_by_xpath("//span[@class='author notFaded']")
        author.append(autore.text)
    except:
        author.append('error')
    
    #isbn
    try:
        time.sleep(2)
        isbn = driver.find_elements_by_xpath("//ul[@class='a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list']/li/span[@class='a-list-item']")
        code = isbn[4].get_attribute('textContent')
        isbn_code.append(code[16:30])
    except:
        isbn_code.append('error')
        
    
    #data pubb
    try:
        date = isbn[0].get_attribute('textContent')
        pub_date.append(date)
    except:
        pub_date.append('error')
        
    
    #lingua
    try:
        lang = isbn[1].get_attribute('textContent')
        lingua.append(lang)
    except:
        lingua.append('error')
    
    
    #numero pagine
    try:
        try:
            pages = isbn[2].get_attribute('textContent')
            p=re.findall(r'\d+', pages)
            n_pages.append(p[0])
        
        except IndexError:  #capita che ci sia un codiice ASIN che sballa le cose
            pages = isbn[3].get_attribute('textContent')
            p=re.findall(r'\d+', pages)
            n_pages.append(p[0])
    except:
        n_pages.append('error')
    
   
    #stars
    try:
        stars = driver.find_element_by_xpath("//div[@id='reviewsMedley']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/div[@class='a-section a-spacing-none a-spacing-top-mini cr-widget-ACR']/div[@class='a-fixed-left-grid AverageCustomerReviews a-spacing-small']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/i[@data-hook='average-star-rating']/span[@class='a-icon-alt']")
        stars = stars.get_attribute('textContent')
        avg_rat.append(float(stars[0:3].replace(',','.')))
    except:
        avg_rat.append('error')
    
    
    #distributed rating
    try:
        dist_rat = driver.find_element_by_xpath("//div[@id='reviewsMedley']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/span[@class='cr-widget-Histogram']")
        dist_rat = dist_rat.get_attribute('textContent')
        distr_rating.append(re.findall(r'\d+\%' ,dist_rat))
    except:
        distr_rating.append('error')


    
    #numero voti e recensioni 
    try:
        time.sleep(1)
        all_rev=driver.find_element_by_xpath("//a[@data-hook='see-all-reviews-link-foot']")
        n_rev_link=all_rev.get_attribute('href')
        url2=str(n_rev_link)
        driver.get(url2)
        n_rev=driver.find_element_by_xpath("//div[@data-hook='cr-filter-info-review-rating-count']/span")
        v_r=n_rev.text.split('|')
        count=0
        for el in v_r:
            el=el.replace('.','')
            if count==0:
                n_voti.append(float(re.findall(r'\d+' ,el)[0]))
                count+=1
            else:
                n_review.append(float(re.findall(r'\d+' ,el)[0]))
    except:
        n_voti.append('error')
        n_review.append('error')

#print(len(gen))
print(len(title))
print(len(author))
print(len(isbn_code))
print(len(n_pages))
print(len(avg_rat))
print(len(distr_rating))
print(len(n_voti))
print(len(n_review))
print(len(pub_date))
print(len(lingua))

```

```python
#adesso creo delle liste che attacco al dataset iniziale


sh_ita['Title_az']=title
sh_ita['Author_az']=author
sh_ita['ISBN_az']=isbn_code
sh_ita['N_Pages_az']=n_pages
sh_ita['Rating_az']=avg_rat
sh_ita['Distr_Rating_az']=distr_rating
sh_ita['Num_Voti_az']=n_voti
sh_ita['Num_Review_az']=n_review
sh_ita['Pub_Date_az']=pub_date
sh_ita['Lingua_az']=lingua

```

```python
autor2=[]
for el in author:
    if el[-10:-1] == ' (Autore)':
        autor2.append(el[:-10])
    elif el[-9:] == ' (Autore)':
        autor2.append(el[:-9])
#decidiamo di conservare i traduttor
    else:
        autor2.append(el)
        
sh_ita['Author_az']=autor2
sh_ita
    
    
#sistemiamo l'autore per amazon cosi da poterne fare confronto  
```

```python

aut=0
err=0
diff_aut=[]
for index,row in sh_ita.iterrows():
    aut_ano=row['Author'].replace('.','').replace(',','').lower()
    aut_ama=row['Author_az'].replace('.','').replace(',','').lower()
    if aut_ano == aut_ama:
        aut+=1
    elif aut_ano.split(' ')[1] == aut_ama.split(' ')[0]: #casi tipo: william gibson vs gibson william
        aut+=1
            
    elif row['Author_az'] == 'error':
        err+=1
        
    else:
        diff_aut.append((aut_ano,aut_ama))
print(aut)
print(err)
diff_aut

```

```python
#al netto degli errori su amazon, in circa 400 casi viene riportato il traduttore piuttosto che l'autore originale probabilmente
#1/4 degli autori è diverso dagli originali. 
#infatti spesso si vedono nomi italiani, in quanto il link anobii rimanda all'edizione italiana
```

```python
from fuzzywuzzy import fuzz

tit=0
er=0
diff_tit=[]
for index,row in sh_ita.iterrows():
    tit_ano=row['Title'].replace('.','').replace(',','').lower()
    tit_ama=row['Title_az'].replace('.','').replace(',','').lower()
    tit_val=fuzz.ratio(tit_ano, tit_ama)
    if tit_val>65:
        tit+=1
    #elif aut_ano.split(' ')[1] == aut_ama.split(' ')[0]: #casi tipo: william gibson vs gibson william
        #aut+=1
            
    elif tit_ama == 'error':
        er+=1
        
    else:
        diff_tit.append((tit_ano,tit_ama))
print(tit)
print(er)
diff_tit
```

```python
#quasi 2/3 dei libri hanno similarità del 65% nel titolo, dato un mero confornto di stringa
```

```python
isb=0
err_ano=0
err_ama=0
diff_isb=[]
for index,row in sh_ita.iterrows():
    isb_ano=row['ISBN_an']
    isb_ama=row['ISBN_az']
    if str(isb_ano) == str(isb_ama).replace('-',''):
        isb+=1
            
    elif isb_ano == 'error' or len(isb_ano)!=13:
        err_ano+=1
    elif isb_ama == 'error' or len(isb_ano)!=14:
        err_ama+=1
        
    else:
        diff_isb.append((isb_ano,isb_ama))
print(isb)
print(err_ano)
print(err_ama)
diff_isb

```

```python
#non ci sono isbn diversi a livello di numero, dettato dal fatto che quello che si raccoglie è tutto errore praticmanete
#si evincono solo 340 match per gli isbn
#molte edizioni vecchie non hanno isbn a 13 cifre ad esempio 
```
