```python
#creazione dataframe

amazon_reviews=pd.DataFrame(list(zip(titoli,isbn_code,genere,userlist,ratings,rev_date,rev_place,n_likes,rev_content)), columns=['Title','ISBN','Genere','Username','Stars','Date','ReviewCountry','N_Likes','Content'])
#amazon_ita=amazon_ita.reset_index()
#affil_covid_def=affil_covid_def.drop('index',axis=1)
amazon_en_reviews=amazon_reviews[amazon_reviews['Content']!= 'error'].reset_index()
amazon_en_reviews
```

```python
ama_en_nodup=amazon_en_reviews.drop_duplicates( subset=['Title','Content','Username'],keep='first',inplace=False).reset_index()
ama_en_nodup
```

```python
#check reviews inglesi

x=0
ind=[]
for el in ama_en_nodup['Content']:
    if type(el) is not str:
        ind.append(x)
        #print(el)
    x+=1
print(len(ind))
#sono tutte stringhe, quindi tutto ok 

from langdetect import detect
lang=[]
for index, row in ama_en_nodup['Content'].iteritems():
    try:
        language = detect(row) #detecting each row
        lang.append(language)
    except:
        lang.append('not found')
lang[:10]

print(len(lang))


ama_en_nodup['Language_AZ']=lang
ama_en_nodup

```

```python
amazon_engrev=ama_en_nodup[ama_en_nodup['Language_AZ'] == 'en'].drop('level_0',axis=1).reset_index()
amazon_engrev
```

```python
#check errori

amazon_engrev=amazon_engrev[amazon_engrev['Content']!='error']
amazon_engrev=amazon_engrev[amazon_engrev['Content']!='']
amazon_engrev=amazon_engrev[amazon_engrev['Content']!=' ']
amazon_engrev=amazon_engrev[amazon_engrev['Content']!='[]']
amazon_engrev=amazon_engrev[amazon_engrev['Content']!='no_link']
amazon_engrev
```

```python
amazon_engrev=amazon_engrev.sort_values(by=['Title','Date'],ascending=False)
amazon_engrev
```

```python
#pulizia finale
amazonen_fin=amazon_engrev.drop(['level_0','index','Language_AZ'],axis=1).reset_index()
amazonen_fin=amazonen_fin.drop('index',axis=1)
amazonen_fin
```
