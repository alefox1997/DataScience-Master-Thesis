```python
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
PATH='C:\\Users\\Fossati\\Desktop\\Tesi_Ale\\chromedriver91.exe'
driver = webdriver.Chrome(PATH)
```

```python
#INTEGRAZIONE SHARED BOOKS ITALIA
```

```python
#importo il file di AMAZON IT ed estraggo la lista degli ISBN (o dei titoli se con ISBN non riesci)

sh_ita=pd.read_csv("\\Users\\Fossati\\Desktop\\Tesi_Ale\\Dati\\Dati_Italia\\Shared_IT.csv")
#sh_ita=amazon_ita.drop('Unnamed: 0',axis=1)
sh_ita
```

```python
#controllo duplicati nei titoli, perchÃ¨ ci sono piÃ¹ edizioni ovviamente
sh_ita=sh_ita.drop_duplicates( subset=['Title'],keep='first',inplace=False)
sh_ita=sh_ita.reset_index() #yes24 filtrato dalle righe duplicate
sh_ita
```

```python
num_pag=[]
stelle=[]
num_voti=[]
num_rece=[]
#in aggiunta:
pub_date=[]
lingua=[]
sug_gen=[]
lnk_ama=[]
```

```python
anobii_lnk=sh_ita['Link_Anobii'].tolist()
no_lnk=0
for lnk in anobii_lnk:

    time.sleep(2)
    count=0
    #otteniamo il link con ISBN
    try:
        try:
            driver.get(lnk)
            count+=1
        except:
            time.sleep(3)
            driver.get(lnk)
            count+=1
    except:
        num_pag.append('no link')
        num_voti.append('no link')
        num_rece.append('no link')
        stelle.append('no link')
        pub_date.append('no link')
        lingua.append('no link')
        sug_gen.append('no link')
        lnk_ama.append('no link')
        no_lnk+=1
        
    

        
    
    if count==1:
    #ora che siamo dentro la pagina delle statistiche:
    
    #link amazon
        try:
            time.sleep(3)
            linkama=driver.find_element_by_xpath("//div['_ngcontent-serverapp-c117']/app-side-list['_ngcontent-serverapp-c117']/div['_ngcontent-serverapp-c181']/div['_ngcontent-serverapp-c181']/ul/a").get_attribute('href')
            lnk_ama.append(linkama)
        except:
            try:
                time.sleep(12)
                driver.find_element_by_xpath("//div['_ngcontent-serverapp-c117']/app-side-list['_ngcontent-serverapp-c117']/div['_ngcontent-serverapp-c181']/div['_ngcontent-serverapp-c181']/ul/a").get_attribute('href')
                lnk_ama.append(linkama)
            except:
                lnk_ama.append('error')
                
            
    #numero pagine:
        try:
            pag=driver.find_elements_by_xpath("//app-about-edition['_ngcontent-serverapp-c117']['_nghost-serverapp-c186']/div[@class='ng-star-inserted']/div['_ngcontent-serverapp-c186']")
            n_pag=pag[5].text
            n_pag=int(re.findall(r'\d+' ,n_pag)[0])
            num_pag.append(n_pag)
            time.sleep(1)
        except:
            num_pag.append('error')
            
    #pubdate 
        try:
            #pag=driver.find_elements_by_xpath("//app-about-edition['_ngcontent-serverapp-c117']['_nghost-serverapp-c186']/div[@class='ng-star-inserted']/div['_ngcontent-serverapp-c186']")
            date=pag[2].text
            pub_date.append(date)
            time.sleep(1)
        except:
            pub_date.append('error')
    
    #lingua
        
        try:
            #pag=driver.find_elements_by_xpath("//app-about-edition['_ngcontent-serverapp-c117']['_nghost-serverapp-c186']/div[@class='ng-star-inserted']/div['_ngcontent-serverapp-c186']")
            lang=pag[0].text
            lingua.append(lang)
            time.sleep(1)
        except:
            lingua.append('error')
            
    #genere suggerito:
        aus_gen=[]
        try:
            g=driver.find_elements_by_xpath("//app-genres-edition['_ngcontent-serverapp-c117 _nghost-serverapp-c191']/div[@class='row col']/a")
            if len(g)!=0: #ci sono + generi
                for el in g:
                    aus_gen.append(el.get_attribute('textContent'))
                sug_gen.append(aus_gen)
            
            else: #sarÃ  solo 1 genere
                g2=driver.find_element_by_xpath("//app-genres-edition['_ngcontent-serverapp-c117 _nghost-serverapp-c191']/div[@class='row col']/a")
                sug_gen.append(g2.get_attribute('textContent'))
        except:#non ci sarÃ  genere
            sug_gen.append('error')
    
    #numero voti
        try:
            time.sleep(3)
            n_stars=driver.find_element_by_xpath("//div[@class='expandible']/span[@class='star-count']/span[@class='color-yellow']").text
            try:
                n_stars=n_stars.replace('.','')
                num_voti.append(int(n_stars))
            except:
                n_stars=n_stars.replace(',','')
                num_voti.append(int(n_stars))                
            #time.sleep(1)
        except:
            num_voti.append('error')
    
    #numero recensioni
        try:
            try:
                time.sleep(1)
                n_rev=driver.find_element_by_xpath("//app-number-text['_ngcontent-serverapp-c117']/a[@class='ng-star-inserted']").text
                try:
                    n_rev=n_rev.replace('.','')
                    num_rece.append(int(n_rev))
                except:
                    n_rev=n_rev.replace(',','')
                    num_rece.append(int(n_rev))                   
                #time.sleep(1)
            except:
                n_rev=driver.find_element_by_xpath("//app-number-text['_ngcontent-serverapp-c117']/span[@class='no-anchor ng-star-inserted']").text
                try:
                    n_rev=n_rev.replace('.','')
                    num_rece.append(int(n_rev))
                except:
                    n_rev=n_rev.replace(',','')
                    num_rece.append(int(n_rev))
        except:
            num_rece.append('error')
            
    
    #stelle
        try:
            stars=driver.find_element_by_xpath("//div[@class='expandible']/ngb-rating['_ngcontent-serverapp-c178']").get_attribute('aria-valuenow')
            stelle.append(round(float(stars),2))
        #time.sleep(1)
        except:
            stelle.append('error')
    
    #teniamo conto errori ad ogni ciclo
    print(no_lnk)
    print(len(num_rece))
    er=0
    for el in num_rece:
        if el=='error':
            er+=1
    print(er)
    print('\n')
    


```
