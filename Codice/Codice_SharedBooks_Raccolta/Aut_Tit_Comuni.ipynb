```python
#lista autori in comune: 
#eliminaimo duplicati dalle liste e facciamo match tra liste delle 3 lingue

aut_en=meta_en['Author'].to_list()
aut_it=meta_it['Author'].to_list()
aut_ko=meta_ko['Author'].to_list()

#drop duplicati dalle liste
aut_en2=[]
aut_it2=[]
aut_ko2=[]

def dropdup(lista1, lista2):
    for el in lista1:
        if el not in lista2:
            lista2.append(el)
    return lista2

dropdup(aut_en,aut_en2)
dropdup(aut_it,aut_it2)
dropdup(aut_ko,aut_ko2)
        
```

```python
print(len(aut_en2))
print(len(aut_it2))
print(len(aut_ko2))  #si nota come gli autori siano circa dimezzati dopo aver eliminato i duplicati per le 3 lingue
```

```python
#quali sono in comune?

print(set(aut_en2).intersection(aut_it2, aut_ko2) ) #quali
print(len(set(aut_en2).intersection(aut_it2, aut_ko2))) #quanti

#32 autori comuni alle 3 lingue

#qui sotto si trova la lista degli autori comuni allle 3 lingue
```

```python
tup_en=[]
tup_ko=[]
tup_it=[]

def tupla(aut,tit,newlist):
    i=0
    while i < len(aut):
        tup = (aut[i],tit[i])
        print(tup)
        newlist.append(tup)
        i+=1
    return newlist

tupla(aut_en, tit_en, tup_en)
tupla(aut_ko, tit_ko, tup_ko)
tupla(aut_it, tit_it2, tup_it)
```

```python
def fuzzy_tup(lang1,lang2):
    diz2={}
    for el in lang1:
        for tit in lang2:
            value = fuzz.ratio(el,tit)
            if value > 80:
                chiave=(el, tit)
                diz2[chiave]=value
    return diz2
```
