```python
goodreads=[]
autori_good=[]
isbn_gr=[]
no_lang=[] #lista per capire per quali indici non è stato possibile controllare la lingua
lnk='https://www.goodreads.com/shelf/show/'
for autore in aut_list:
    time.sleep(3)
    autore2=autore.replace(' ','%20')
    link= lnk + autore2
    
    for i in range(1,15,1):
        time.sleep(3)
        url= link + "?page=" + str(i)
        
        try: #perchè può essere che sfori di pagina
            driver.get(url)
            driver.find_element_by_class_name('authorName') #è la prova per scorrere le pagine in cui non ci sono libri
        except:
            break
            
        #passiamo alla ricerca dei link
        links=driver.find_elements_by_xpath("//div[@class='elementList']/div[@class='left']/a")
        links2=[] 
        #autori_good=[]
        for el in links: 
            obj=el.get_attribute('href')
            links2.append(obj)
            
        #filtro duplicati da link
        links3=[]
        #count=0
        for el in links2: 
            if el not in links3:
                links3.append(el)
 
        #appendiamo alla lista goodreads principale 
        c=0
        for elem in links3:
            time.sleep(2)
            driver.get(elem)#entriamo per prendere l'edizione inglese
            try:
                time.sleep(1)
                lan=driver.find_element_by_xpath("//div[@id='bookDataBox']/div[@class='clearFloats']/div[@class='infoBoxRowItem'][@itemprop='inLanguage']").text
                if lan == 'English':
                    goodreads.append(elem)
                    aut_name=driver.find_element_by_class_name('authorName').text
                    autori_good.append(aut_name)
                    #isbn
    
                    try:
                        isbn=driver.find_element_by_xpath("//div[@class='buttons']/div[@id='bookDataBox']/div[@class='clearFloats']/div[@class='infoBoxRowItem']/span[@class='greyText']/span[@itemprop='isbn']")
                        isbn_gr.append(int(isbn.get_attribute('textContent')))
                    except:
                        isbn_gr.append('error')


                
                    
            except: #except si intende come: se non trovi la lingua fa nulla, porta dentro i dati lo stesso
                try:
                    time.sleep(3)
                    driver.get(elem)
                    lan=driver.find_element_by_xpath("//div[@id='bookDataBox']/div[@class='clearFloats']/div[@class='infoBoxRowItem'][@itemprop='inLanguage']").text
                    if lan == 'English':
                        goodreads.append(elem)
                        try:
                            aut_name=driver.find_element_by_class_name('authorName').text
                            autori_good.append(aut_name)
                        except:
                            autori_good.append('error')
                        #isbn
    
                        try:
                            isbn=driver.find_element_by_xpath("//div[@class='buttons']/div[@id='bookDataBox']/div[@class='clearFloats']/div[@class='infoBoxRowItem']/span[@class='greyText']/span[@itemprop='isbn']")
                            isbn_gr.append(int(isbn.get_attribute('textContent')))
                        except:
                            isbn_gr.append('error')
                            
                except:
                    time.sleep(3)
                    driver.get(elem)
                    no_lang.append(c)
                    try:
                        aut_name=driver.find_element_by_class_name('authorName').text
                        autori_good.append(aut_name)
                    except:
                        autori_good.append('error')
                    
                    goodreads.append(elem)
                    #isbn
    
                    try:
                        isbn=driver.find_element_by_xpath("//div[@class='buttons']/div[@id='bookDataBox']/div[@class='clearFloats']/div[@class='infoBoxRowItem']/span[@class='greyText']/span[@itemprop='isbn']")
                        isbn_gr.append(int(isbn.get_attribute('textContent')))
                    except:
                        isbn_gr.append('error')
                        
                c+=1
                
            #driver.get(url)
            
            
print(len(goodreads))   
print(len(autori_good))   
print(len(isbn_gr))    
```

```python


#per riassumere, abbiamo preso tutte le opere per oguno dei 32 autori da goodreads, filtrando per lingua inglese ed eliminando
#quelle con autore non presente nella lista


title=[]
for index,row in gr_data3.iterrows():
    num=re.findall(r'\d+' ,row['Link_Goodreads'])[0]
    titolo=row['Link_Goodreads'].split(num)[-1]
    title.append(titolo[1:])
print(len(title))
print(title[:25])

gr_data3['Title']=title
gr_data3=gr_data3.drop(['index','level_0'],axis=1)
gr_data3

gr_data3=gr_data3.sort_values(by=['Author'])
gr_data3=gr_data3.reset_index()
gr_data3

```

```python
#schema codice:
# scorri il dataset e provi a creare il link dall'isbn, altrimenti lo crei dal titolo (rivaluta bontà del link però IT)
# nota: link vuoti vengono presi, quindi manda una ricerca e crea errore se li incontri, a quel punto generi link
# prendi top 5 e cerchiamo quale è il libro migliore, come?
# calcoliamo similarità per il titolo e per l'autore e sommiamo i punteggi, vedendo chi tra i top 5 ha similarity maggiore
# edizione inglese?
# di quello estraiamo il link
```

```python
#AMAZONEN

#posso pensare di prendere gli stessi che ho preso da goodreads sfruttando l'ISBN o il titolo

from fuzzywuzzy import fuzz

amazon_en=[] #lista dei link diretti ai libri
link='https://www.amazon.it/s?k=9780099408390&i=stripbooks&__mk_it_IT=%C3%85M%C3%85%C5%BD%C3%95%C3%91&ref=nb_sb_noss'
pre=link[:26]
post=link[39:]
mylinks=[] #lista link pagina contente più libri
#count=0

for index,row in gr_data3.iterrows():
    if row['ISBN_gr'] != 'error' and len(str(row['ISBN_gr']))==13:
        isbn=row['ISBN_gr']
        lnk=pre+str(isbn)+post
        #qui valutiamo se il link con l'isbn sia funzionante, perchè spesso non fornisce risultati. Se è sbagliato creiamo
        #con il titiolo
        try: 
            time.sleep(2)
            driver.get(lnk)
            time.sleep(2)
            driver.find_element_by_class_name("a-size-medium.a-color-base.a-text-normal")
            mylinks.append(lnk)
        except:
            tit=row['Title']
            lnk=pre+str(tit)+post
            mylinks.append(lnk)

    else:
        tit=row['Title']
        lnk=pre+str(tit)+post
        mylinks.append(lnk)
        
        
    #una volta scelto il link da utilizzare, entriamo e facciamo le operazioni
    try:
        time.sleep(2)
        driver.get(lnk)
    except:
        try:
            time.sleep(5)
            driver.get(lnk)
        except:
            amazon_en.append('no link')
            continue
    
    #cerchiamo di capire quale sia la sezione con maggiore similarity per autore e titolo, quindi raccogliamo autori e titoli
    
    try:
        time.sleep(3)
        tit_aus=driver.find_elements_by_class_name("a-size-medium.a-color-base.a-text-normal") #titoli
        time.sleep(3)
        aut_aus=driver.find_element_by_class_name("a-row") #sezione autori
    except:
        amazon_en.append('no link')
        continue
    
    
    
    c=0
    aus=[]
    try: #se i libri in pagina presentati sono 5
        while c<5:
            #calcoliamo i punteggi di similarity
            tit_val=fuzz.ratio(tit_aus[c].text, row['Title'])
            if '|' in aut_aus[c].text:
                aut_val=fuzz.ratio(aut_aus[c].text.split('|')[1][4:-1] , row['Author'])
            else:
                aut_val=fuzz.ratio(aut_aus[c].text.split(',')[0][3:] , row['Author'])
                
            somma= tit_val + aut_val #calcoliamo la somma delle similarities e appendiamole alla lista esterna
           
            aus.append(int(somma))
            c+=1
        max_value = max(aus)
        max_index = aus.index(max_value)
        
        
        
    
    except: #se i libri sono 3
        try:
            while c<3:
            #calcoliamo i punteggi di similarity
                tit_val=fuzz.ratio(tit_aus[c].text, row['Title'])
                if '|' in aut_aus[c].text:
                    aut_val=fuzz.ratio(aut_aus[c].text.split('|')[1][4:-1] , row['Author'])
                else:
                    aut_val=fuzz.ratio(aut_aus[c].text.split(',')[0][3:] , row['Author'])
                
                somma= tit_val + aut_val #calcoliamo la somma delle similarities e appendiamole alla lista esterna
           
                aus.append(int(somma))
                c+=1
            max_value = max(aus)
            max_index = aus.index(max_value)
            
        except:  #nel caso estremo in cui ne hai meno di 3
        
            max_index=0

    try:
        time.sleep(3)
        final_lnk=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-micro']/div[@class='a-row a-size-small']/span['aria-label']/a")
    except:
        amazon_en.append('no link')
        continue
    
    try:
        time.sleep(3)
        url=str(final_lnk[max_index].get_attribute('href'))
        amazon_en.append(url)
        #driver.get(url) #si acquisice l'url per lo scraping a sto punto
        #flag+=1
    except:
        amazon_en.append('no link')
        continue        
        







```
