```python
from fuzzywuzzy import fuzz

amazon_en=[]
for index,row in sh_eng.iterrows():
    lnk=row['Link_ama_prov']
    try:
        time.sleep(1)
        driver.get(lnk)
    except:
        try:
            time.sleep(30) #per battere il controllo
            driver.get(lnk)
        except:
            amazon_en.append(lnk)
            continue
    
    #cerchiamo di capire quale sia la sezione con maggiore similarity per autore e titolo, quindi raccogliamo autori e titoli
    
    try:
        time.sleep(3)
        driver.find_element_by_class_name("a-size-medium.a-color-base.a-text-normal") #prova da superare per capire in che pagina siamo
        tit_aus=driver.find_elements_by_class_name("a-size-medium.a-color-base.a-text-normal") #titoli
        time.sleep(3)
        aut_aus=driver.find_element_by_class_name("a-row") #sezione autori
    except:
        amazon_en.append(lnk)
        continue
    
    
    
    c=0
    aus=[]
    try: #se i libri in pagina presentati sono 5
        while c<5:
            #calcoliamo i punteggi di similarity
            tit_val=fuzz.ratio(tit_aus[c].text, row['Title'])
            if '|' in aut_aus[c].text:
                aut_val=fuzz.ratio(aut_aus[c].text.split('|')[1][4:-1] , row['Author'])
            else:
                aut_val=fuzz.ratio(aut_aus[c].text.split(',')[0][3:] , row['Author'])
                
            somma= tit_val + aut_val #calcoliamo la somma delle similarities e appendiamole alla lista esterna
           
            aus.append(int(somma))
            c+=1
        max_value = max(aus)
        max_index = aus.index(max_value)
        
        
        
    
    except: #se i libri sono 3
        try:
            while c<3:
            #calcoliamo i punteggi di similarity
                tit_val=fuzz.ratio(tit_aus[c].text, row['Title'])
                if '|' in aut_aus[c].text:
                    aut_val=fuzz.ratio(aut_aus[c].text.split('|')[1][4:-1] , row['Author'])
                else:
                    aut_val=fuzz.ratio(aut_aus[c].text.split(',')[0][3:] , row['Author'])
                
                somma= tit_val + aut_val #calcoliamo la somma delle similarities e appendiamole alla lista esterna
           
                aus.append(int(somma))
                c+=1
            max_value = max(aus)
            max_index = aus.index(max_value)
            
        except:  #nel caso estremo in cui ne hai meno di 3
        
            max_index=0

    try:
        time.sleep(3)
        final_lnk=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-micro']/div[@class='a-row a-size-small']/span['aria-label']/a")
    except:
        amazon_en.append(lnk)
        continue
    
    try:
        time.sleep(3)
        url=str(final_lnk[max_index].get_attribute('href'))
        amazon_en.append(url)
        #driver.get(url) #si acquisice l'url per lo scraping a sto punto
        #flag+=1
    except:
        amazon_en.append(lnk)
        continue        
        


```
