```python
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
PATH='C:\\Users\\Fossati\\Desktop\\Tesi_Ale\\chromedriver91.exe'
driver = webdriver.Chrome(PATH)
```

```python
#INTEGRAZIONE SHARED BOOKS GOODREADS



#importo il file di AMAZON IT ed estraggo la lista degli ISBN (o dei titoli se con ISBN non riesci)

sh_eng_good=pd.read_csv("\\Users\\Fossati\\Desktop\\Tesi_Ale\\Dati\\Dati_English\\sh_en_goodreads.csv")
#sh_ita=amazon_ita.drop('Unnamed: 0',axis=1)
sh_eng_good







```

```python
title=[]
author=[]
isbn_code=[]
n_pages=[]
avg_rat=[]
distr_rating=[]
n_voti=[]
n_review=[]
pub_date=[]
lingua=[]



#genere suggerito lo prendiamo da anobii per disambiguare

ama=-1 #contatore per richiamare l'altra lista
lista_lnk2=sh_eng_good['Link_Amazon'].tolist()
lista_ama=sh_eng_good['Link_2_Amazon'].tolist()

for link in lista_lnk2:
    ama+=1
    time.sleep(3)
    try:
        try:
            driver.get(link)
        except:
            time.sleep(15)
            driver.get(link)
    except:
        title.append('error')
        author.append('error')
        isbn_code.append('error')
        n_pages.append('error')
        avg_rat.append('error')
        distr_rating.append('error')
        n_voti.append('error')
        n_review.append('error')
        pub_date.append('error')
        continue
        
    
    time.sleep(3)
    
    try:
        driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/h1[@id='title']/span[@id='productTitle']")
        #il titolo funge da prova in quanto ci sono alcuni link da amazon che non funzionano
    except:
        try:
            time.sleep(5)
            driver.get(lista_ama[ama]) #se c'è errore sfruttiamo il link catturato con mio codice
        except:
            title.append('error')
            author.append('error')
            isbn_code.append('error')
            n_pages.append('error')
            avg_rat.append('error')
            distr_rating.append('error')
            n_voti.append('error')
            n_review.append('error')
            pub_date.append('error')
            continue
            
            
            
            
    #titolo
    
    
    try:
        titolo = driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/h1[@id='title']/span[@id='productTitle']")
        title.append(titolo.text)
    except: #potrebbbe esserci il sistema di controllo scraping. Poi re scrapi quelli che non hanno titolo
        try:
            time.sleep(15)
            titolo = driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/h1[@id='title']/span[@id='productTitle']")
            title.append(titolo.text)
        except:
            time.sleep(1)
            title.append('error')
    
    #autore
    try:
        autore = driver.find_element_by_xpath("//span[@class='author notFaded']")
        author.append(autore.text)
    except:
        author.append('error')
    
    #isbn
    try:
        time.sleep(2)
        isbn = driver.find_elements_by_xpath("//ul[@class='a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list']/li/span[@class='a-list-item']")
        code = isbn[4].get_attribute('textContent')
        isbn_code.append(code[16:30])
    except:
        isbn_code.append('error')
        
    
    #data pubb
    try:
        date = isbn[0].get_attribute('textContent')
        pub_date.append(date)
    except:
        pub_date.append('error')
        
    
    #lingua
    try:
        lang = isbn[1].get_attribute('textContent')
        lingua.append(lang)
    except:
        lingua.append('error')
    
    
    #numero pagine
    try:
        try:
            pages = isbn[2].get_attribute('textContent')
            p=re.findall(r'\d+', pages)
            n_pages.append(p[0])
        
        except IndexError:  #capita che ci sia un codiice ASIN che sballa le cose
            pages = isbn[3].get_attribute('textContent')
            p=re.findall(r'\d+', pages)
            n_pages.append(p[0])
    except:
        n_pages.append('error')
    
   
    #stars
    try:
        stars = driver.find_element_by_xpath("//div[@id='reviewsMedley']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/div[@class='a-section a-spacing-none a-spacing-top-mini cr-widget-ACR']/div[@class='a-fixed-left-grid AverageCustomerReviews a-spacing-small']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/i[@data-hook='average-star-rating']/span[@class='a-icon-alt']")
        stars = stars.get_attribute('textContent')
        avg_rat.append(float(stars[0:3].replace(',','.')))
    except:
        avg_rat.append('error')
    
    
    #distributed rating
    try:
        dist_rat = driver.find_element_by_xpath("//div[@id='reviewsMedley']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/span[@class='cr-widget-Histogram']")
        dist_rat = dist_rat.get_attribute('textContent')
        distr_rating.append(re.findall(r'\d+\%' ,dist_rat))
    except:
        distr_rating.append('error')


    
    #numero voti e recensioni 
    try:
        time.sleep(1)
        all_rev=driver.find_element_by_xpath("//a[@data-hook='see-all-reviews-link-foot']")
        n_rev_link=all_rev.get_attribute('href')
        url2=str(n_rev_link)
        driver.get(url2)
        n_rev=driver.find_element_by_xpath("//div[@data-hook='cr-filter-info-review-rating-count']/span")
        v_r=n_rev.text.split('|')
        count=0
        for el in v_r:
            el=el.replace('.','')
            if count==0:
                n_voti.append(float(re.findall(r'\d+' ,el)[0]))
                count+=1
            else:
                n_review.append(float(re.findall(r'\d+' ,el)[0]))
    except:
        n_voti.append('error')
        n_review.append('error')

#print(len(gen))
print(len(title))
print(len(author))
print(len(isbn_code))
print(len(n_pages))
print(len(avg_rat))
print(len(distr_rating))
print(len(n_voti))
print(len(n_review))
print(len(pub_date))
print(len(lingua))
















```

```python
#adesso creo delle liste che attacco al dataset iniziale


sh_eng_good['Title_az']=title #per valutare anche affidabilità dei link
sh_eng_good['Author_az']=author
sh_eng_good['ISBN_az']=isbn_code
sh_eng_good['N_Pages_az']=n_pages
sh_eng_good['Rating_az']=avg_rat
sh_eng_good['Distr_Rating_az']=distr_rating
sh_eng_good['Num_Voti_az']=n_voti
sh_eng_good['Num_Review_az']=review
sh_eng_good['Lingua_az']=pub_date
sh_eng_good['Lingua_az']=lingua

```
