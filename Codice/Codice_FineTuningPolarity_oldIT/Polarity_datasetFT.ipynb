```python
def length(dataset,colname):
    rev=dataset[colname].tolist()
    totfrasi=[]
    totparole=[]
    for review in rev:
        perpunto=review.replace("!",".").replace("?",".").replace("|",".").split(".")
        totfrasi.append(len(perpunto))
        
        parole=review.split(" ")
        totparole.append(len(parole))
        
        frasi_per_rev=int(sum(totfrasi)/len(totfrasi))
        parole_per_frasi=int(sum(totparole)/len(totparole))
        
    return print(frasi_per_rev),print(parole_per_frasi)
```

```python
#funzione per splittare train e test
from random import sample
def train_test_spl(dataset):
    train=dataset.sample(int(len(dataset)*0.6))
    test=dataset.drop(train.index.tolist(),axis=0).sample(int(len(dataset)*0.3))
    dev=dataset.drop(train.index.tolist(),axis=0).drop(test.index.tolist(),axis=0)
    return train,test,dev

```
