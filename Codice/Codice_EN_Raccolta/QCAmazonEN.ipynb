```python
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
PATH='C:\\Users\\aless\\OneDrive\\Desktop\\Laurea_magistrale\\Tesi\\chromedriver.exe'
driver = webdriver.Chrome(PATH)
```

```python
isbn_list=MetaGR_EN['ISBN'].tolist()
```

```python
link='https://www.amazon.it/s?k=9780099408390&i=stripbooks&__mk_it_IT=%C3%85M%C3%85%C5%BD%C3%95%C3%91&ref=nb_sb_noss'
pre=link[:26]
post=link[39:]
mylinks=[]
#count=0

for isbn in isbn_list:
    lnk=pre+str(isbn)+post
    mylinks.append(lnk)  

mylinks[:20]
```

```python
#creiamo liste vuote per poter poi fare i dataframe

avg_rat=[]
distr_rating=[]
n_voti=[]
n_review=[]
```

```python
for lnk in mylinks: 
    flag=0
    time.sleep(3)
    
    try:
        driver.get(lnk)
    except:
        avg_rat.append('no link')
        distr_rating.append('no link')
        n_voti.append('no link')
        n_review.append('no link')
        continue
    
    #codice per riuscire a prendere il libro di interesse. capita che ci siano molte edizioni e in testa alla pagina amazon ci sia quella meno recensita, non originale
    # cerchiamo l'edizione del libro con cui gl utenti hanno maggiornamente interagito
    
    try:
        n_rev_aus=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-micro']/div[@class='a-row a-size-small']/span['aria-label']/a[@class='a-link-normal']/span")
    except:
        avg_rat.append('no link')
        distr_rating.append('no link')
        n_voti.append('no link')
        n_review.append('no link')
        continue
        
        
    c=0
    aus=[]
    try: #se i libri in pagina presentati sono 5
        for el in n_rev_aus[:5]:
            rep=n_rev_aus[c].text.replace('.','')
            aus.append(int(rep))
            c+=1
        max_value = max(aus)
        max_index = aus.index(max_value)
    
    except: #se i libri sono 3
        try:
            for el in n_rev_aus[:3]:
                rep=n_rev_aus[c].text.replace('.','')
                aus.append(int(rep))
                c+=1
            max_value = max(aus)
            max_index = aus.index(max_value)
        except:
        
            max_index=0

    try:
        n_rev_lnk=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-spacing-top-micro']/div[@class='a-row a-size-small']/span['aria-label']/a")
    except:
        avg_rat.append('no link')
        distr_rating.append('no link')
        n_voti.append('no link')
        n_review.append('no link')
        continue
    
    try:
        url=str(n_rev_lnk[max_index].get_attribute('href'))
        driver.get(url) #si acquisice l'url per lo scraping a sto punto
        flag+=1
    except:
        avg_rat.append('no link')
        distr_rating.append('no link')
        n_voti.append('no link')
        n_review.append('no link')
        continue
    
    
    
    #una volta appreso se esiste il link e preso lo stesso, iniziamo con lo scraping vero e proprio
    
    if flag==1:

    
    #stars
        time.sleep(1)
        try:
            stars = driver.find_element_by_xpath("//div[@id='reviewsMedley']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/div[@class='a-section a-spacing-none a-spacing-top-mini cr-widget-ACR']/div[@class='a-fixed-left-grid AverageCustomerReviews a-spacing-small']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/i[@data-hook='average-star-rating']/span[@class='a-icon-alt']")
            stars = stars.get_attribute('textContent')
            avg_rat.append(float(stars[0:3].replace(',','.')))
        except:
            avg_rat.append('error')
    
    #distributed rating
        time.sleep(1)
        try:
            dist_rat = driver.find_element_by_xpath("//div[@id='reviewsMedley']/div[@class='a-fixed-left-grid-inner']/div[@class='a-fixed-left-grid-col a-col-left']/span[@class='cr-widget-Histogram']")
            dist_rat = dist_rat.get_attribute('textContent')
            distr_rating.append(re.findall(r'\d+\%' ,dist_rat))
        except:
            distr_rating.append('error')

    #numero voti e recensioni 
        try:
            time.sleep(1)
            all_rev=driver.find_element_by_xpath("//a[@data-hook='see-all-reviews-link-foot']")
            n_rev_link=all_rev.get_attribute('href')
            url2=str(n_rev_link)
            driver.get(url2)
            n_rev=driver.find_element_by_xpath("//div[@data-hook='cr-filter-info-review-rating-count']/span")
            v_r=n_rev.text.split('|')
            count=0
            for el in v_r:
                el=el.replace('.','')
                if count==0:
                    n_voti.append(float(re.findall(r'\d+' ,el)[0]))
                    count+=1
                else:
                    n_review.append(float(re.findall(r'\d+' ,el)[0]))
        except:
            n_voti.append('error')
            n_review.append('error')
    
```
